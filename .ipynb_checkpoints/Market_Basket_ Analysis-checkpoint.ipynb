```python
#To get all column names into a list
cols = data.columns
```

<span style="color:red">To get a better understanding of the data, each column of the dataset was analysed. This is to visualize the most frequent item bought with the item in the column.</span>

```python
from collections import Counter

#Loops through all columns
for col in cols:
    #Counts the different items in the columns
    letter_counts = Counter(list(data[col]))
    #Makes the different items and item count into a dataframe
    df = pd.DataFrame.from_dict(letter_counts, orient='index')
    #sets the dimension of the graph
    plt.rcParams["figure.figsize"] = 15,6
    #plots a bar graph for the item count in each column
    df.plot(kind='bar')
    #Adds the column name to the corresponding graph
    plt.title(col)
```
