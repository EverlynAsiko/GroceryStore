```python
cols = data.columns
```

```python
from collections import Counter

for col in cols:
    letter_counts = Counter(list(data[col]))
    df = pd.DataFrame.from_dict(letter_counts, orient='index')
    plt.rcParams["figure.figsize"] = 15,6
    df.plot(kind='bar')
```
